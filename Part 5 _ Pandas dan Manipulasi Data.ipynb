```python
# mengimport pandas ke notebook

import pandas as pd
```

```python
# membaca data dengan read_csv

df = pd.read_csv('SuperStore.csv')
```

```python
# menampilkan 5 baris teratas

df.head()
```

## Mengubah tipe data menjadi tanggal

```python
# mengubah order date dan ship date menjadi tipe data tanggal

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
# menampilkan tipe data dari df yaitu DataFrame

type(df)
```

```python
# kolom kolom dari DataFrame adalah Series

type(df['Customer_ID'])
```

```python
# menampilkan daftar kolom dari df

df.columns
```

```python
# memilih kolom tertentu dari df

df['Category']
```

```python
# memilih lebih dari satu kolom

nama_kolom = ['Category','Customer_ID','City']
df[nama_kolom]
```

```python
# slicing menggunakan loc
# syntax: df.loc[baris,kolom]
# pada loc, cara memilih kolom adalah dengan memanggil nama kolom tersebut

# menampilkan 10 baris pertama dari kolom Order_ID
df.loc[:10,'Order_ID']
```

```python
# menampilkan 5 baris teratas dari nama-nama
# kolom yang telah didefinisikan

nama_kolom = ['Category','Customer_ID','City']
df.loc[:5,nama_kolom]
```

```python
# pada list kolom df, kolom dengan index 4 adalah 'Sales'

df.columns[4]
```

```python
# pada iloc, kolom dipanggil dengan nomor indexnya

df.iloc[:10,4]
```

```python
# memanggil 10 baris pertama dari
# kolom dengan index 1,3,5,7

df.iloc[:10,[1,3,5,7]]
```

```python
# Untuk melakukan filtering, gunakan syntax: df[kondisi]

# contoh, memfilter sales dengan nilai lebih dari 100

df[df['Sales'] > 100].head()
```

```python
# memfilter dengan dua atau lebih kondisi
# ingat: untuk menghindari kesalahan penulisan ekspresi, gunakan tanda kurung

# gunakan bitwise operator
# & untuk and
# | untuk or
# ~ untuk not

# memfilter transaksi dengan sales > 100 dari kategori Furniture
df[(df['Sales'] > 100) & (df['Category'] == 'Furniture')].head()
```

```python
df[~(df['Category']=='Furniture')].head()
```

```python
df[(df['City']=='Henderson') | (df['City']=='Los Angeles')].head()
```

## Sorting

```python
# sorting berdasarkan tanggal
# secara default, ascending = False
# sehingga diurutkan dari tanggal terkecil ke terbesar

df.sort_values(by='Order_Date')
```

```python
df.sort_values(by='Order_Date', ascending=False)
```

```python
# sorting berdasarkan dua kolom

df.sort_values(by=['Category','Sub-Category'])
```

```python
# Sorting menggunakan dua kolom
# Category secara menurun
# Sub-Category secara menaik

df.sort_values(by=['Category','Sub-Category'], ascending=[False, True])
```

## Iterasi

```python
# iterasi dengan list of columns

# karena df.columns menghasilkan list of columns
# maka df.columns dapat digunakan sebagai iterator

# iterasi berdasarkan kolom berguna saat kita ingin
#mengaplikasikan operasi berulang terhadap kolom tertentu

for kolom in df.columns:
  print(kolom)
```

```python
# disamping itu, iterasi juga dapat dilakukan terhadap index
df.index
```

```python
# contoh
for idx in df.index:
  print(df['Order_ID'][idx], df['Order_Date'][idx])
```

```python
# iterasi baris per baris dengan iterrows()

for index, row in df.iterrows():
  print(row['Order_ID'], row['Order_Date'])
```
